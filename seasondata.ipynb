```python
lakers = [team for team in nba_teams if team["full_name"] == "Los Angeles Lakers"][0]
lakers
```

```python
lakers23 = teamyearbyyearstats.TeamYearByYearStats(team_id = lakers['id'])
lakers23.get_data_frames()[0]
```
